```python
sales_data_cleaned = sales_data.dropna(subset=['customer_id'])
```

```python
sales_data_cleaned['age'].fillna(sales_data_cleaned['age'].median(), inplace = True)
```

```python
# prompt: clean na name with 'Unknown'

sales_data_cleaned['name'].fillna('Unknown', inplace = True)
```

```python
sales_data_cleaned['age'].fillna(sales_data_cleaned['age'].median(), inplace = True)
```

```python
sales_data_cleaned
```

```python
sales_data_cleaned['email'].fillna('unknown@gmail.com', inplace = True)
```

```python
sales_data_cleaned
```

```python
sales_data_cleaned['purchase_date'].fillna(sales_data_cleaned['purchase_date'].median(), inplace = True)
```

```python
sales_data_cleaned
```

```python
sales_data_cleaned['amount'].fillna(sales_data_cleaned['amount'].mean(), inplace = True)
```

```python
sales_data_cleaned
```

```python
# prompt: clean age <20 with mode

mode_age = sales_data_cleaned[sales_data_cleaned['age'] >= 20]['age'].mode()[0]
sales_data_cleaned.loc[sales_data_cleaned['age'] < 20, 'age'] = mode_age
sales_data_cleaned
```
